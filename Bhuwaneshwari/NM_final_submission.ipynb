```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Actual Hubble value (m.Gyr^-1)
H0 = 67400 / (3.09 * 1e22) * (3600 * 24 * 365 * 1e9)
```

```python
# Initial Omega parameters
Omega1_m = 0.3
Omega1_r = 1e-4
Omega1_l = 0.0
Omega1_k = 1 - Omega1_m - Omega1_r - Omega1_l

Omega2_m = 0.3
Omega2_r = 1e-4
Omega2_l = 0.7
Omega2_k = 1 - Omega2_m - Omega2_r - Omega2_l

Omega3_m = 5.0
Omega3_r = 1e-4
Omega3_l = 0.0
Omega3_k = 1 - Omega3_m - Omega3_r - Omega3_l

Omega4_m = 1.0
Omega4_r = 1e-4
Omega4_l = 0.0
Omega4_k = 1 - Omega4_m - Omega4_r - Omega4_l
```

```python
# Differential system function
def syseq(y, t, Omega_m, Omega_r, Omega_l, Omega_k):
    # Actual Hubble value
    H0 = 67400 / (3.09 * 1e22) * (3600 * 24 * 365 * 1e9)
    return [y[1],(-(H0**(2)/2)*(Omega_m/y[0]**(3)+2*Omega_r/y[0]**(4)-2*Omega_l))*Omega_m*(1/(H0**(2))*y[1]**(2)-Omega_r/y[0]**(2)-Omega_l*y[0]**(2)-Omega_k)**(-1)]
```

```python
# Initial scale factor and derivated scale factor
a1_0 = 1.0
a1_prim_0 = H0 * ((Omega1_m/a1_0) + (Omega1_r/(a1_0 ** 2)) + Omega1_k + Omega1_l * (a1_0**2))**(0.5)
init1 = [a1_0, a1_prim_0]

a2_0 = 1.0
a2_prim_0 = H0 * ((Omega2_m/a2_0) + (Omega2_r/(a2_0 ** 2)) + Omega2_k + Omega2_l * (a2_0**2))**(0.5)
init2 = [a2_0, a2_prim_0]

a3_0 = 1.0
a3_prim_0 = H0 * ((Omega3_m/a3_0) + (Omega3_r/(a3_0 ** 2)) + Omega3_k + Omega3_l * (a3_0**2))**(0.5)
init3 = [a3_0, a3_prim_0]

a4_0 = 1.0
a4_prim_0 = H0 * ((Omega4_m/a4_0) + (Omega4_r/(a4_0 ** 2)) + Omega4_k + Omega4_l * (a4_0**2))**(0.5)
init4 = [a4_0, a4_prim_0]
```

```python
# Model 1 : 
```

```python
# Starting and ending time
t_begin = [-1.0e-5, 1e-5]
t_final = [-15, 22]

# Solving ODEs in two parts: backwards and forwards
for i in range(2):
    # Differential systems solving
    t1 = np.linspace(t_begin[i], t_final[i], 5000)
    y1 = odeint(syseq, init1, t1, args=(Omega1_m, Omega1_r, Omega1_l, Omega1_k), rtol=1e-10, atol=1e-10)
    plt.figure(0)
    plt.plot(t1, y1[:, 0], 'b')

# Plot settings
plt.xlabel('Cosmic Time in Gyr')
plt.ylabel('Relative size of the universe')
plt.title('Hyperbolic geometry (k=−1)')
plt.grid(True)
plt.ylim([0, 4])
plt.savefig('hyperbolic_geometry(k=−1)_final.png')  # Save the plot as a PNG file
plt.show()
```

```python
# Starting and ending time
t_begin = [-1.0e-5, 1e-5]
t_final = [-14, 22]

# Solving ODEs in two parts: backwards and forwards
for i in range(2):
    # Differential systems solving
    t1 = np.linspace(t_begin[i], t_final[i], 5000)
    t2 = t1.copy()
    y2 = odeint(syseq, init2, t2, args=(Omega2_m, Omega2_r, Omega2_l, Omega2_k), rtol=1e-10, atol=1e-10)
    plt.figure(1)
    plt.plot(t2, y2[:, 0], 'r')

# Plot settings
plt.xlabel('Cosmic Time in Gyr')
plt.ylabel('Relative size of the universe')
plt.title('ΛCDM model')
plt.grid(True)
plt.ylim([0, 4])

plt.savefig('LambdaCDM_model_final.png')  # Save the plot as a PNG file
plt.show()
```
